```python
import os
def load_vocab():
    file_path = os.path.join('aclImdb', 'imdb.vocab')
    f = open(file_path, 'r')
    lines = f.readlines()
    f.close()
    vocab = set()
    for line in lines:
        word = line.strip('\n')
        vocab.add(word)
    return vocab
```

```python
vocab_set = load_vocab()
vocab_set.add('<UNK>')
vocab_set.add('<PAD>')
words_set = sorted(vocab_set)
words_index = {x: i for i, x in enumerate(vocab_set)}
```

```python
def load_data(flag):
    file_path = os.path.join('aclImdb', flag)
    res = []
    labels_dict = {'neg': 0, 'pos': 1}
    sentences = []
    labels = []
    for label in labels_dict.keys():
        for file_name in os.listdir(os.path.join(file_path, label)):
            with open(os.path.join(os.path.join(file_path, label), file_name), encoding='utf-8') as f:
                sentence = f.readline()
                sentence = sentence.lower()
                words_temp = sentence.split(' ')
                words = []
                for word in words_temp:
                    if word in vocab_set:
                        words.append(word)
                    else:
                        words.append('<UNK>')
                sentences.append(words)
                labels.append(labels_dict[label])
    return sentences, labels
```

```python
train_sentences, train_labels = load_data('train')
test_sentences, test_labels = load_data('test')
```

```python
new_train=[]
new_test=[]
for i in range(len(train_sentences)):
  new_train.append(' '.join(train_sentences[i]))
  new_test.append(' '.join(test_sentences[i]))
  if i==0:
    print(train_sentences[i])
    print(new_train[i])
    print(type(new_train[i]))
  # test_sentences[i]=' '.join(test_sentences[i])
  
```

```python
#文本特征向量转化模块
from sklearn.feature_extraction.text import CountVectorizer
vec = CountVectorizer()
X_train = vec.fit_transform(new_train)
X_test = vec.transform(new_test)
 
```

```python
#Bayes
from sklearn.naive_bayes import MultinomialNB
mnb = MultinomialNB()
mnb.fit(X_train,train_labels)
y_predict = mnb.predict(X_test)
from sklearn.metrics import classification_report
print('The accruacy of Naive Bayes is ',mnb.score(X_test,test_labels))
print(classification_report(test_labels,y_predict,target_names=['o','n']))
```

```python
# 3_gram
vec2 = CountVectorizer(ngram_range=(1,3))
X_train2 = vec2.fit_transform(new_train)
X_test2 = vec2.transform(new_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
mnb = MultinomialNB()
mnb.fit(X_train2,train_labels)
y_predict = mnb.predict(X_test2)
from sklearn.metrics import classification_report
print('The accruacy of Naive Bayes is ',mnb.score(X_test2,test_labels))
print(classification_report(test_labels,y_predict,target_names=['o','n']))
```

```python
#KNN
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier

knc = KNeighborsClassifier()
knc.fit(X_train,train_labels)
y_predict_k = knc.predict(X_test)
 
print('The Accruacy of kNN is',knc.score(X_test, test_labels))
from sklearn.metrics import classification_report
print(classification_report(test_labels,y_predict_k,target_names=['o','n']))
```

```python
#Decision Tree
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
dtc.fit(X_train,train_labels)
y_predict_d = dtc.predict(X_test)

print (dtc.score(X_test, test_labels))
print(classification_report(test_labels,y_predict_d,target_names=['o','n']))
```
